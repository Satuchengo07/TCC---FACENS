```python
abfss = spark.conf.get('spark.saneamento.datalake.default')
```

```python
tabela = 'macro2963'
```

```python
df = spark.sql(f"""
               select lrc.periodo as periodo,
               lrc.idligacao as ligacao,
               c.categoria as categoria,
               lrc.qtdeconomia as qtdeconomia,
               lrc.idunidadeoperacional as idunidadeoperacional,
               current_date as igest_date
from std_san.ligacaoresumocategoria lrc
inner join std_san.categoria c on (lrc.idcategoria = c.id and c.idunidadeoperacional = lrc.idunidadeoperacional)
inner join std_san.periodo p on (lrc.idperiodo = p.id and p.idunidadeoperacional = lrc.idunidadeoperacional)
where lrc.idunidadeoperacional = 23""")

display(df.limit(10))
```

```python
df.write.mode("overwrite").format("delta").option("mergeSchema", "true").option("path", f"{abfss}/refined/comercial/{tabela}").saveAsTable(f"ref_comercial.{tabela}")
```
